## 1.a 
Give a function F that takes a single argument n. This function should print a sequence of n strings as described in the examples below: <br>
Ex: **n = 3**

--A--<br>
-BAB-<br>
CBABC<br>

**n = 4**

---A---<br>
--BAB--<br>
-CBABC-<br>
DCBABCD

```python

def F(n):
    def s(n):
        result = ''  # Initialize an empty string
        for i in range(n):
            result += chr(65 + i)  # add the characters
        return result  

    def Row(x):
        # create the mirrored string (without repeating the center character)
        mirror = x + x[-2::-1]  
        return mirror

    # now the loop to generate the pattern
    for i in range(1, n + 1):
        base_seq = s(i)
        
        mirrored_seq = Row(base_seq)
        dashes = "-" * (n - i)
        print(f"{dashes}{mirrored_seq}{dashes}")

# Tests
F(10)
F(6)
```

## 1.b
Make a function F that takes only one argument, a dictionary(dict) d.
The keys of d are integers and the values of d are a tuple of type (x (int), y (int)).
You must print out the dict in the format "-key-, -x-, -y-" with each entry in a new line. Print it for each of the three sorted orders, by key values ascending, by x values descending, by y values ascending.

```python
def F(d : dict):
    # key values ascending
    print("Sorted by key (ascending):")
    for key in sorted(d.keys()):
        x, y = d[key]
        print(f"{key}, {x}, {y}")

    # x values descending
    print("\nSorted by x (descending):")
    for key, (x, y) in sorted(d.items(), key=lambda item: item[1][0], reverse=True):
        print(f"{key}, {x}, {y}")

    # y values ascending
    print("\nSorted by y (ascending):")
    for key, (x, y) in sorted(d.items(), key=lambda item: item[1][1]):
        print(f"{key}, {x}, {y}")

# Tests
F({1 : (1, 2), 2 : (-1, 4), 5 : (-4, 3), 4 : (2, 3)})
F({-8 : (4, 2), 6 : (-3, 4), 7 : (2, 1), 5 : (9, -10)})
```

#### 2.b: Print out the total credits and calculte CPI of each student. 
CPI is the weighted average of core courses, and electives (weights being the letter grades converted to number AP,AA=10, AB=9, BB=9, BC=7, CC=6)

```python
with open("student_records.csv", "r") as file:
    rows = file.readlines()  # Read all lines into a list
    header = rows[0].strip().split(",")  # First row as header
    student_data = [dict(zip(header, row.strip().split(","))) for row in rows[1:]]  # Map header to row values

def calculate_cpi_and_credits(student_data):
    """
    Calculates total credits and CPI for each student.
    """
    grade_points = {"AP": 10, "AA": 10, "AB": 9, "BB": 8, "BC": 7, "CC": 6}
    roll_to_data = {}

    for record in student_data:
        roll = record["roll_number"]
        if roll not in roll_to_data:
            roll_to_data[roll] = {"total_credits": 0, "weighted_grades": 0, "course_count": 0}
        
        credit = int(record["credit"])
        grade = record["grade"]
        
        if grade in grade_points:
            weight = grade_points[grade]
            roll_to_data[roll]["total_credits"] += credit
            roll_to_data[roll]["weighted_grades"] += weight * credit
            roll_to_data[roll]["course_count"] += 1

    for roll, data in roll_to_data.items():
        total_credits = data["total_credits"]
        weighted_grades = data["weighted_grades"]
        cpi = round(weighted_grades / total_credits, 2) if total_credits > 0 else 0
        print(f"Roll Number: {roll}, Total Credits: {total_credits}, CPI: {cpi}")

calculate_cpi_and_credits(student_data)
```

#### 2.c: Print out the names of all students who meet the graduation requirements 
Atleast 20 credist of core course, 15 credits of department elective, 10 credits of flexible elective and 5 credits of hasmed electives

```python
with open("student_records.csv", "r") as file:
    rows = file.readlines()  # Read all lines from the file

student_data = [row.strip().split(",") for row in rows[1:]]  # Skip the header and split each row

def graduation_requirements(student_data):
        
    # Organize by student
    students = {}
    for row in student_data:
        roll_number, department, credit, course_name, course_type, grade = row
        credit = int(credit)  # Convert credit to integer
        
        if roll_number not in students:
            students[roll_number] = []

        students[roll_number].append({
            "credit": credit,
            "course_type": course_type,
            "grade": grade
        })

    
    graduates = []
    for roll_number, courses in students.items():
        core_credits = sum(course["credit"] for course in courses if course["course_type"] == "core")
        department_elective_credits = sum(course["credit"] for course in courses if course["course_type"] == "department_elective")
        felxible_elective_credits = sum(course["credit"] for course in courses if course["course_type"] == "felxible_elective")
        hasmed_credits = sum(course["credit"] for course in courses if course["course_type"] == "hasmed_elective")

        if (core_credits >= 20 and 
            department_elective_credits >= 15 and 
            felxible_elective_credits >= 10 and 
            hasmed_credits >= 5):
            graduates.append(roll_number)

    print("Students who meet graduation requirements:")
    for roll_number in graduates:
        print(f"  Roll Numbers: {roll_number}")


graduation_requirements(student_data)
```

## Part A
Minimize the function $f(x, y) = 2(x - y - 3)^2 + 4(x + 2y + 1)^4$.<br>
With the constraints : $ x - y \ge -3, (x + 2)^2 + (y + 1)^2 \le 5$ <br>
Using scipy.optimize.minimize with constraints. (Hint: Look at the examples in the official documentation)

```python
%pip install numpy scipy
import numpy as np
from scipy.optimize import minimize

def f(x):
    return 2*(x[0] - x[1] - 3)**2 + 4*(x[0] + 2*x[1] + 1)**4

# Define the constraints
def constraint1(x):
    return x[0] - x[1] + 3  # x - y >= -3, or x - y + 3 >= 0

def constraint2(x):
    return 5 - (x[0] + 2)**2 - (x[1] + 1)**2  # (x + 2)^2 + (y + 1)^2 <= 5

# Initial guess for the optimization variables
x0 = [0, 0]

# Define the constraints in the format needed for minimize
con1 = {'type': 'ineq', 'fun': constraint1}  # Linear constraint: x - y + 3 >= 0
con2 = {'type': 'ineq', 'fun': constraint2}  # Nonlinear constraint: (x + 2)^2 + (y + 1)^2 <= 5

result = minimize(f, x0, constraints=[con1, con2])

print(f"Optimal solution: x = {result.x[0]}, y = {result.x[1]}")
print(f"Objective function value: {result.fun}")
print(f"Success: {result.success}")
print(f"Message: {result.message}")
```

# Part B
Evaluate the line integral of the function $f(x, y) = x^2 + y^4$ along the circle $ x^2 + y^2 = 3 $ anticlockwise (scalar integral, not vector). You must use scipy for integration but you may use parameterization

```python
import numpy as np
from scipy.integrate import quad

# Define the integrand
def integrand(t):
    return (3 * np.cos(t)**2 + 9 * np.sin(t)**4) * np.sqrt(3)

# Perform the integration over the interval [0, 2*pi]
result, error = quad(integrand, 0, 2 * np.pi)

print(f"Line integral result: {result}")
print(f"Estimated error: {error}")
```

```python
### TODO 2.1
### If init_centers is None, initialize the centers by selecting K data points at random without replacement
### Else, use the centers provided in init_centers
### Return : np array of size Kx2

def initialise_centers(data, K, init_centers=None):
    if init_centers is None:
        indices = np.random.choice(data.shape[0], K, replace=False) #random, no replacement
        centers = data[indices]
    else:
        centers = np.array(init_centers)
    
    print("Initial Centers:\n", centers)
    return centers
K=2
initialise_centers(data, K, init_centers=None)
```

```python
### TODO 2.2
### Initialize the labels to all ones to size (N,) where N is the number of data points
### Return : np array of size N
def initialise_labels(data):
    N = data.shape[0]
    
    labels = np.ones(N, dtype=int)
    
    return labels

initialise_labels(data)
```

```python
### TODO 3.1 : E step
### For Each data point, find the distance to each center
### Return : np array of size NxK
centers = initialise_centers(data, K)
data = np.loadtxt(data_path, delimiter=',')
def calculate_distances(data, centers):
        differences = data[:, np.newaxis, :] - centers[np.newaxis, :, :]
        distances = np.sqrt(np.sum(differences**2, axis=2))
    
        return distances
distances = calculate_distances(data, centers)
print("Distances:\n", distances)
```

```python
### TODO 3.2 : E step
### For Each data point, assign the label of the nearest center
### Return : np array of size N
def update_labels(distances):
    labels = np.argmin(distances, axis=1)
    return labels

data = load_data('spice_locations.txt')
K = 5 #playing around with this value
centers = initialise_centers(data, K)
distances = calculate_distances(data, centers)
labels = update_labels(distances)

print("Labels:", labels)
```

```python
### TODO 4 : M step
### Update the centers to the mean of the data points assigned to it
### Return : np array of size Kx2
def update_centers(data, labels, K):
    new_centers = np.zeros((K, data.shape[1]))  

    for k in range(K):
        cluster_points = data[labels == k]
        
        if cluster_points.size > 0:  
            new_centers[k] = np.mean(cluster_points, axis=0)

    return new_centers
update_centers(data, labels, K)
```

```python
### TODO 6 : Check convergence
### Check if the labels have changed from the previous iteration
### Return : True / False
def check_termination(labels1, labels2):
    def check_termination(labels1, labels2):
        return np.array_equal(labels1, labels2) #yes or no if they converged or not
```

```python
### DON'T CHANGE ANYTHING IN THE FOLLOWING FUNCTION
def kmeans(data_path:str, K:int, init_centers):
    '''
    Input :
        data (type str): path to the file containing the data
        K (type int): number of clusters
        init_centers (type numpy.ndarray): initial centers. shape = (K, 2) or None
    Output :
        centers (type numpy.ndarray): final centers. shape = (K, 2)
        labels (type numpy.ndarray): label of each data point. shape = (N,)
        time (type float): time taken by the algorithm to converge in seconds
    N is the number of data points each of shape (2,)
    '''
    data = load_data(data_path)    
    centers = initialise_centers(data, K, init_centers)
    labels = initialise_labels(data)

    start_time = time.time() # Time stamp 

    while True:
        distances = calculate_distances(data, centers)
        labels_new = update_labels(distances)
        centers = update_centers(data, labels_new, K)
        if check_termination(labels, labels_new): break
        else: labels = labels_new
 
    end_time = time.time() # Time stamp after the algorithm ends
    return centers, labels, end_time - start_time 
```
